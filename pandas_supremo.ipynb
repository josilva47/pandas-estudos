```python
import pandas as pd
import numpy as np

df_vendas = pd.read_csv("vendas_tech.csv", encoding='utf-8', low_memory=False)
display(df_vendas)
df_gerentes = pd.read_excel("gerentes_lojas.xlsx")
display(df_gerentes)
```

```python
# display(df_vendas.head(10)) 
# display(df_vendas.tail(10))
# display(df_vendas.sample(7))
# display(df_vendas.shape)
# display(df_vendas.columns)
display(df_vendas.info())
# display(df_vendas.describe())

```

```python
#TRATAMENTO DE DADOS

# display(df_vendas['Loja']) #Seleciona uma coluna do DataFrame
# df_vendas[['Loja', 'Produto']] #Seleciona varias colunas
df_analise = df_vendas.drop(columns=['Data_Base']) #Exlui coluna
# df_analise = df_analise.dropna() #Exclui todas as linhas que estao com valores nulos
df_analise['Loja'] = df_analise['Loja'].fillna('Online') #Preenche valores onde o campo estiver vazio
df_analise['Data'] = pd.to_datetime(df_analise['Data'], format="%Y-%m-%d") #Converte para Data 
df_analise['Loja'] = df_analise['Loja'].str.strip() #Remove os espaços inuteis
df_analise['Loja'] = df_analise['Loja'].str.title() #A primeira letra de cada palavra fica em Maiusculo
df_analise = df_analise.drop_duplicates(subset=['ID_Pedido']) #Remove as linhas duplicadas por coluna especifica

display(df_analise.info())

```

```python
#CRIAR NOVAS COLUNAS

#Faturamento
df_analise['Faturamento'] = df_analise['Preco_Unitario'] * df_analise['Qtd'] #Cria a coluna com uma conta simples

#Forma de Venda
df_analise['Forma_de_Venda'] = np.where(df_analise['Loja'] == 'Online', 'Online', 'Fisica') #Cria a coluna por meio de uma condiçao (Condiçao, valor True, valor False)


#Regiao
display(df_analise['Loja'].unique()) #Retorna valores unicos em uma coluna

dic_regioes = {
    'São Paulo': 'Sudeste', 
    'Belo Horizonte': 'Sudeste',
    'Online': 'Online', 
    'Rio De Janeiro': 'Sudeste',
    'Salvador': 'Nordeste', 
    'Recife': 'Nordeste', 
    'Curitiba': 'Sul', 
    'Porto Alegre': 'Sul'
}

df_analise['Regiao'] = df_analise['Loja'].map(dic_regioes)

display(df_analise)
display(df_analise.isna().sum()) #Verifica se tem alguma linha de todas as colunas vazias
```

```python
#ANALISAE -> FILTRAR

df_analise = df_analise.sort_values(by=['Data', 'Faturamento']) #Ordena por coluna (ascending=False -> decrescente)
df_analise = df_analise.reset_index(drop=True) #Reseta o indice e exclue o que ja tinha (Depois de filtros/ordenação, o índice fica bagunçado.)

#Filtra por indice da linha e nome da coluna 
loja = df_analise.loc[3, 'Loja'] 
produto = df_analise.loc[3, 'Produto']
cliente = df_analise.loc[3, 'Cliente']
print(loja, produto, cliente)

#Filtra por posiçao do indice e da coluna
loja = df_analise.iloc[3, 2]
produto = df_analise.iloc[3, 3]
cliente = df_analise.iloc[3, 6]
print(loja, produto, cliente)

#Filtrar por condicional
df_vendas_sp = df_analise[df_analise['Loja']=='São Paulo']
df_vendas_sp.to_csv('Vendas_SP.csv', index=False) #Exporta os dados sem o indice
display(df_vendas_sp)

df_vendas_2024 = df_analise[df_analise['Data'].dt.year==2024] #Extrai o ano de uma coluna de data (A coluna precisa ser datetime)
display(df_vendas_2024)

#Duas Condições
df_vendas_HDMI_SUL = df_analise[(df_analise['Produto']=='Cabo HDMI') & (df_analise['Regiao']=='Sul')]
display(df_vendas_HDMI_SUL)
```

```python
#ANALISE POR GRUPAMENTO
analise_lojas = df_analise[['Loja', 'Faturamento']].groupby('Loja').sum() #AGRUPA dados e faz cálculos em cima desses grupos
analise_lojas = analise_lojas.sort_values(by='Faturamento', ascending=False)
analise_lojas = analise_lojas.reset_index()
analise_lojas['Faturamento'] = analise_lojas['Faturamento'].map('R${:,.2f}'.format)

display(analise_lojas)
```
